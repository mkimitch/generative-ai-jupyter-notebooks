# Lecture Notes: Understanding BERT

## Introduction to BERT

BERT (Bidirectional Encoder Representations from Transformers) is a large language model trained on 3.3 billion words, with 2.5 billion words from Wikipedia and 0.8 billion words from BooksCorpus. BERT is designed to understand the context of words in a sentence in a way that mimics human comprehension, making it highly effective for various natural language processing tasks.

## BERT Training Components

### 1. Masked Language Model (MLM)

BERT uses a bidirectional training process called the Masked Language Model. This process involves the following steps:

- **Masking Words:** In any given sentence, 15% of the words are randomly masked.
  - **Original Sentence:** I want to learn how to cook yummy vegan curries.
  - **Masked Sentence:** I want to [MASKED] how to cook yummy vegan curries.

- **Predicting Masked Words:** BERT looks at the context of the masked word from both sides (before and after the masked word) to predict the masked word.

This method allows BERT to train effectively and understand the English language and the words used within it.

### 2. Next Sentence Prediction (NSP)

BERT also employs a Next Sentence Prediction task during training to understand the relationship between sentence pairs. This involves:

- **Comparing Sentence Pairs:** Determining if the first sentence is logically followed by the second sentence.
  - **Correct Pair Example:** I love to cook and entertain. I have been experimenting with vegan curries.
  - **Random Pair Example:** I love to cook and entertain. The bird is flapping away.

During training, 50% of the sentence pairs are correct, and 50% are random.

## Input and Embeddings

BERT's inputs are the sum of three types of embeddings:

1. **Token Embeddings:** Represent individual words or parts of words using WordPiece embeddings, breaking down words into smaller units called tokens. Each token is assigned a unique numerical representation to capture its semantic meaning.
   
2. **Position Embeddings:** Indicate the position of each word in the sentence, helping BERT understand the order of words.

3. **Segment Embeddings:** Identify different parts of the input text, indicating whether the tokens belong to the same sentence or different sentences.

Source: [BERT Paper](https://arxiv.org/abs/1810.04805)

## Output of BERT

BERT's base model provides contextualized representations of the input text at both the token level and the sentence level:

1. **Token-level Output:** For each input token, BERT produces a vector representation capturing the semantic meaning based on the entire input sequence context. This output is useful for tasks like named entity recognition, part-of-speech tagging, and sentiment analysis.

2. **Sentence-level Output:** BERT also provides a pooled representation of the entire input sequence, used for tasks like text classification. For instance, in our case, it helps predict 22 categories of diagnosis based on the text description of symptoms.

## Model Architecture

BERT consists of multiple layers of transformer blocks. Each block includes a self-attention mechanism and feedforward neural networks. These blocks allow BERT to capture hierarchical and contextual representations of the input text.

## Parameter-Efficient Fine-Tuning

### What is Parameter-Efficient Fine-Tuning?

Parameter-efficient fine-tuning adapts a pre-trained model to a specific task, minimizing the number of additional parameters introduced during fine-tuning. Techniques include:

- Keeping some layers of the pre-trained model frozen.
- Reusing pre-trained layers without modification, except for the task-specific final layers.

### LORA (Low-Rank Adaptation) Method

For fine-tuning our model, we will use the LORA method. This involves:

- **Maintaining Pre-Trained Weights:** The weight matrices in each of the 12 attention layers are kept frozen.
- **Training Additional Matrices:** Two smaller matrices are trained and incorporated into the pre-trained weights to form the final weight matrix used for training the Query and Key components of the attention layers.

This method drastically reduces the number of parameters needing to be trained while allowing fine-tuning of every attention layer.
